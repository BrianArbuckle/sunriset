```python
import sunshine
import datetime
from datetime import timedelta
import math
import pytz

import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

## Creating a Solar Calculations as a Pandas DataFrame using `sunshine.to_pandas()`


```python
lat = 34.0522
long = -118.2437
local_tz = -8

number_of_years = 1
start_date = datetime.date(2019,1,1)

df = sunshine.to_pandas(start_date, lat, long, local_tz, number_of_years)
df
```

## Various plots using the `sunshine` generated DataFrame

```python
df.plot(x='Solar Elevation Angle (degrees)', 
        y='Solar Azimuth Angle (degrees cw from North)', 
        color='orange',
        title='Analemma')
plt.show()
```

```python
plt.plot(df.loc[:, df.columns =='Sunlight Durration (minutes)'], color='orange')
plt.title('Sunlight Durration in Minutes')
plt.ylim(0, 1440)
plt.grid(axis='y', alpha=0.5)

plt.show()
```

```python
plt.plot(df[['Sunrise','Sunset']])
plt.title('Sunrise & Sunset')
plt.legend(df[['Sunrise','Sunset']])
plt.grid(axis='y', alpha=0.5)
plt.show()

# need better y axis labels
```
